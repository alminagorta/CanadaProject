```python
import pandas as pd
import glob, os
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/alminagorta/CanadaProject/master/Benthic_Habitat_Data/2_St_Lawrence_BenthicData.csv'
# Read file into a DataFrame and print its head
xl = pd.read_csv(url)
print(xl.head())
```

```python
#Some Data analysis
#How many classes of species we have?
print('# of macroinvertebrates per site    ',xl['Class'].value_counts())
print('======================')
print('Total # of macroinvertebrates are %s'% xl['Count'].sum())#this to print the output directly
```

```python
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline

#Plotting #bugs per site
sns.catplot(x="Count",y="Class",data=xl,hue="Year")
plt.title('Benthic: %s'%url[0][0]) #here get the name of the file in the plot
#log axes
axes = plt.gca()
#axes.set_ylim([ymin,ymax]) 
axes.set_xscale('log', basex=10)
```
